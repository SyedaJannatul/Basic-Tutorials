```python
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML student age",fontsize=20)
plt.xlabel("age",fontsize=16)
plt.ylabel("no. of students",fontsize=16)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.95,histtype="bar",
         orientation="vertical",color='r',label='ML student')
plt.title("ML student age",fontsize=20)
plt.xlabel("age",fontsize=16)
plt.ylabel("no. of students",fontsize=16)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.95,histtype="bar",
         orientation="vertical",color='r',label='ML student')
plt.hist(py_student_age,bins,rwidth=0.95,histtype="bar",
         orientation="vertical",color='y',label='Py student')
plt.title("ML student age",fontsize=20)
plt.xlabel("age",fontsize=16)
plt.ylabel("no. of students",fontsize=16)
plt.legend(loc=2)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.95,
         histtype="bar",orientation="vertical",color=['r','y'],
         label=['ML student','Py student'])
plt.title("ML student age",fontsize=20)
plt.xlabel("age",fontsize=16)
plt.ylabel("no. of students",fontsize=16)
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(16,9))
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.95,
         histtype="bar",orientation="vertical",color=['r','y'],
         label=['ML student','Py student'])
plt.title("ML student age",fontsize=20)
plt.xlabel("age",fontsize=16)
plt.ylabel("no. of students",fontsize=16)
plt.legend(loc=2)
plt.show()
```
